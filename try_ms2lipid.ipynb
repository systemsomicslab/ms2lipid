```python
import ms2lipid
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

The data to be used must include MSMS spectra, precursor m/z, and identification ID.  



```python
pd.read_csv("https://github.com/systemsomicslab/ms2lipid/raw/main/tests/test_data_csv_neg.csv")[['sampleid', 'MSMSspectrum', 'AverageMz']]
```

Input the **data path**, **measurement ion mode**,column name for **MS/MS spectra** & **precursor m/z** & **identification ID** in the DataFrame   
to perform class prediction.
The data formats supported are CSV, TXT, MSP.

```python
predicted_result_df = ms2lipid.predclass(path="https://github.com/systemsomicslab/ms2lipid/raw/main/tests/test_data_csv_neg.csv",\
                                         ionmode="negative", ms2spc_name="MSMSspectrum", prec_name="AverageMz",ID='sampleid')
```

Check the results
